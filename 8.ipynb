```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

X = np.array([[5,3],
              [10,15],
              [15,12],
              [24,10],
              [30,45],
              [85,70],
              [71,80],
              [60,78],
              [55,52],
              [80,91]])

print("Исходные данные:")
for i, point in enumerate(X):
    print(f"Точка {i+1}: ({point[0]}, {point[1]})")

plt.figure(figsize=(12, 8))

for k in range(2, 5):
    plt.subplot(2, 2, k)
    
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X)
    labels = kmeans.labels_
    centroids = kmeans.cluster_centers_
    
    colors = ['red', 'blue', 'green', 'orange', 'purple']
    
    for i in range(k):
        cluster_points = X[labels == i]
        plt.scatter(cluster_points[:, 0], cluster_points[:, 1], 
                   color=colors[i], label=f'Кластер {i+1}', alpha=0.6)
        plt.scatter(centroids[i, 0], centroids[i, 1], 
                   color=colors[i], marker='X', s=200, linewidths=3)
    
    plt.title(f'K-means, k={k}')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()
    plt.grid(True)
    
    print(f"\nКластеризация с k={k}:")
    print(f"Центроиды:")
    for j, centroid in enumerate(centroids):
        print(f"  Кластер {j+1}: ({centroid[0]:.1f}, {centroid[1]:.1f})")
    print(f"Принадлежность точек к кластерам: {labels}")
    print(f"Инертность: {kmeans.inertia_:.2f}")

inertias = []
K_range = range(1, 6)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)

plt.subplot(2, 2, 1)
plt.plot(K_range, inertias, 'bo-')
plt.xlabel('Количество кластеров (k)')
plt.ylabel('Инерция (Inertia)')
plt.title('Метод локтя')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler

iris = load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names
target_names = iris.target_names

print("Набор данных Ирисы Фишера")
print(f"Размер данных: {X.shape}")
print(f"Количество классов: {len(target_names)}")
print(f"Признаки: {feature_names}")
print(f"Классы: {target_names}")

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

inertias = []
silhouette_scores = []
K_range = range(2, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X_scaled, kmeans.labels_))

plt.figure(figsize=(15, 10))

plt.subplot(2, 3, 1)
plt.plot(K_range, inertias, 'bo-')
plt.xlabel('Количество кластеров (k)')
plt.ylabel('Инерция')
plt.title('Метод локтя')
plt.grid(True)

plt.subplot(2, 3, 2)
plt.plot(K_range, silhouette_scores, 'ro-')
plt.xlabel('Количество кластеров (k)')
plt.ylabel('Silhouette Score')
plt.title('Метод силуэта')
plt.grid(True)

for idx, k in enumerate([2, 3, 4], start=3):
    plt.subplot(2, 3, idx+1)
    
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    labels = kmeans.labels_
    centroids = kmeans.cluster_centers_
    
    colors = ['red', 'blue', 'green', 'orange', 'purple', 'brown', 'pink', 'gray']
    
    for i in range(k):
        cluster_points = X[labels == i]
        plt.scatter(cluster_points[:, 0], cluster_points[:, 1], 
                   color=colors[i], label=f'Кластер {i+1}', alpha=0.6)
    
    plt.scatter(centroids[:, 0], centroids[:, 1], 
               color='black', marker='X', s=200, linewidths=3, label='Центроиды')
    
    plt.title(f'K-means, k={k}')
    plt.xlabel(feature_names[0])
    plt.ylabel(feature_names[1])
    plt.legend()
    plt.grid(True)
    
    print(f"\nКластеризация с k={k}:")
    print(f"Центроидов: {centroids.shape[0]}")
    print(f"Инертность: {kmeans.inertia_:.2f}")
    print(f"Silhouette Score: {silhouette_scores[k-2]:.4f}")

optimal_k = K_range[np.argmax(silhouette_scores)]
print(f"\nОптимальное количество кластеров (по silhouette): {optimal_k}")

kmeans_optimal = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
kmeans_optimal.fit(X_scaled)
optimal_labels = kmeans_optimal.labels_

plt.subplot(2, 3, 6)
for i in range(len(target_names)):
    class_points = X[y == i]
    plt.scatter(class_points[:, 0], class_points[:, 1], 
               color=colors[i], label=target_names[i], alpha=0.6)

plt.title('Истинные классы')
plt.xlabel(feature_names[0])
plt.ylabel(feature_names[1])
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

print("\nПредсказания для 10 случайных точек:")
np.random.seed(42)
random_indices = np.random.choice(len(X), 10, replace=False)

for idx in random_indices:
    point = X[idx]
    true_class = target_names[y[idx]]
    predicted_cluster = kmeans_optimal.predict(scaler.transform([point]))[0]
    print(f"Точка {idx}: {point}")
    print(f"  Истинный класс: {true_class}")
    print(f"  Предсказанный кластер: {predicted_cluster}")
    print()

print("Сравнение кластеризации с истинными классами:")
cluster_distribution = pd.crosstab(optimal_labels, y, 
                                   rownames=['Кластер'], 
                                   colnames=['Истинный класс'])
print(cluster_distribution)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
import pandas as pd
from sklearn.metrics import adjusted_rand_score, normalized_mutual_info_score
from sklearn.preprocessing import LabelEncoder

iris = load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names
target_names = iris.target_names

X_selected = X[:, [0, 2]]

methods = ['single', 'complete', 'average', 'ward']
plt.figure(figsize=(15, 10))

for i, method in enumerate(methods, 1):
    plt.subplot(2, 2, i)
    
    Z = linkage(X_selected, method=method, metric='euclidean')
    
    dendrogram(Z, 
               truncate_mode='lastp',
               p=12,
               leaf_rotation=45.,
               leaf_font_size=12.,
               show_contracted=True)
    
    plt.title(f'Иерархическая кластеризация ({method} linkage)')
    plt.xlabel('Индекс образца')
    plt.ylabel('Расстояние')
    plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

Z = linkage(X_selected, method='ward', metric='euclidean')

plt.figure(figsize=(12, 6))
dendrogram(Z, 
           truncate_mode='lastp',
           p=50,
           leaf_rotation=90.,
           leaf_font_size=10.,
           show_contracted=True)
plt.title('Полная дендрограмма (Ward linkage)')
plt.xlabel('Образцы')
plt.ylabel('Расстояние')
plt.grid(True, alpha=0.3)
plt.show()

num_clusters = 3
clusters = fcluster(Z, num_clusters, criterion='maxclust')

plt.figure(figsize=(12, 8))

colors = ['red', 'blue', 'green', 'orange', 'purple', 'brown', 'pink', 'gray', 'cyan', 'magenta']

for i in range(1, num_clusters + 1):
    cluster_points = X_selected[clusters == i]
    color_idx = (i-1) % len(colors)
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], 
               color=colors[color_idx], 
               label=f'Кластер {i}', alpha=0.7, s=80)

plt.title(f'Иерархическая кластеризация (k={num_clusters})')
plt.xlabel(feature_names[0])
plt.ylabel(feature_names[2])
plt.legend()
plt.grid(True, alpha=0.3)

plt.figure(figsize=(12, 8))
for i, target_name in enumerate(target_names):
    class_points = X_selected[y == i]
    color_idx = i % len(colors)
    plt.scatter(class_points[:, 0], class_points[:, 1], 
               color=colors[color_idx], 
               label=target_name, alpha=0.7, s=80)

plt.title('Истинные классы')
plt.xlabel(feature_names[0])
plt.ylabel(feature_names[2])
plt.legend()
plt.grid(True, alpha=0.3)

plt.show()

print("\nРаспределение точек по кластерам:")
cluster_counts = np.bincount(clusters)
for i in range(1, num_clusters + 1):
    print(f"Кластер {i}: {cluster_counts[i]} точек")

print("\nСравнение с истинными классами:")
comparison_df = pd.DataFrame({
    'Истинный класс': [target_names[label] for label in y],
    'Кластер': clusters
})
cross_tab = pd.crosstab(comparison_df['Истинный класс'], comparison_df['Кластер'])
print(cross_tab)

le = LabelEncoder()
y_encoded = le.fit_transform(y)

ari = adjusted_rand_score(y_encoded, clusters)
nmi = normalized_mutual_info_score(y_encoded, clusters)

print(f"\nМетрики качества кластеризации:")
print(f"Adjusted Rand Index: {ari:.4f}")
print(f"Normalized Mutual Information: {nmi:.4f}")

distance_thresholds = [1, 2, 3, 5, 10]
plt.figure(figsize=(15, 10))

for i, threshold in enumerate(distance_thresholds, 1):
    plt.subplot(2, 3, i)
    
    clusters_threshold = fcluster(Z, threshold, criterion='distance')
    unique_clusters = np.unique(clusters_threshold)
    num_clusters_threshold = len(unique_clusters)
    
    for cluster_idx, cluster_num in enumerate(unique_clusters):
        cluster_points = X_selected[clusters_threshold == cluster_num]
        color_idx = cluster_idx % len(colors)
        plt.scatter(cluster_points[:, 0], cluster_points[:, 1], 
                   color=colors[color_idx], 
                   label=f'Кластер {cluster_num}', alpha=0.6)
    
    plt.title(f'Порог = {threshold}\nКластеров: {num_clusters_threshold}')
    plt.xlabel(feature_names[0])
    plt.ylabel(feature_names[2])
    plt.legend()
    plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
