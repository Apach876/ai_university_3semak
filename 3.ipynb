```python
#1.3.1
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def distance(x,y):
    print(np.linalg.norm(x-y))
    print(np.linalg.norm(x-y) ** 2)
    print(np.linalg.norm(x-y, ord=np.inf))
    print(np.linalg.norm(x-y, ord = 1))


fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')

x = np.array([2,3,0])
y = np.array([7,2,6])
w = np.array([4,8,11])
z = np.array([10,0,0])

distance(x,y)
distance(y,w)
distance(w,z)

ax.scatter(2,3,0)
ax.scatter(7,2,6)
ax.scatter(4,8,11)
ax.scatter(10,0,0)
plt.show()
```
