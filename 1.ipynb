```python
import random
import math
import matplotlib.pyplot as plt


rs = list(range(1, 6))
ls = [random.randint(1, 10) for _ in range(10)]
def f(x):
    return math.sqrt(1 + math.e ** math.sqrt(x) + math.cos(x)**2) / math.fabs(1-math.sin(x)**3) + math.log(2*x)
sl = [f(x) for x in ls]
plt.plot(rs, sl[0:5],  marker='o', linestyle='-', linewidth=2, color='blue', markersize=8)
plt.show()
plt.scatter(rs,sl[5:10])
plt.show()
```
